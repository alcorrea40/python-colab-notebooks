# 🧮 Actividad Interactiva #2: Acceso a columnas y filas en un DataFrame

```python
import pandas as pd
```

## 📊 Creamos un DataFrame de ejemplo

```python
data = {
    'Nombre': ['Ana', 'Luis', 'Carlos', 'María', 'Sofía'],
    'Edad': [23, 35, 45, 32, 28],
    'Ciudad': ['Bogotá', 'Medellín', 'Cali', 'Bogotá', 'Cali'],
    'Ingreso_Mensual': [2500, 3200, 2900, 3100, 3000]
}
df = pd.DataFrame(data)
df
```

## 🔹 Seleccionar columnas

```python
df['Nombre']
```

```python
df[['Nombre', 'Ciudad']]
```

## 🔸 Seleccionar filas por etiqueta con `loc[]`

```python
df.loc[1]
```

```python
df.loc[1:3]
```

## 🔸 Seleccionar filas por posición con `iloc[]`

```python
df.iloc[0]
```

```python
df.iloc[0:3]
```

## 🔄 Combinaciones de selección

```python
df.loc[df['Ciudad'] == 'Bogotá', ['Nombre', 'Edad']]
```
