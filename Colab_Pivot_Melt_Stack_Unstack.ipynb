# Actividad Interactiva #1: Pivot, Melt y Reestructuración con pandas
En este cuaderno vas a practicar cómo transformar la estructura de tus datos usando `pivot()`, `melt()`, `stack()` y `unstack()`.

```python
import pandas as pd
import numpy as np
```

## 1. Crear un DataFrame base para experimentar

```python
df = pd.DataFrame({
    'nombre': ['Ana', 'Luis', 'Ana', 'Luis'],
    'anio': [2020, 2020, 2021, 2021],
    'ventas': [250, 400, 270, 420]
})
df
```

## 2. Aplicar `pivot()` para convertir en formato ancho

```python
df_pivot = df.pivot(index='nombre', columns='anio', values='ventas')
df_pivot
```

## 3. Aplicar `melt()` para volver a formato largo

```python
df_melted = df_pivot.reset_index().melt(id_vars='nombre', var_name='anio', value_name='ventas')
df_melted
```

## 4. Aplicar `stack()` para reorganizar usando índices

```python
df_stacked = df_pivot.stack()
df_stacked
```

## 5. Aplicar `unstack()` para invertir el `stack()`

```python
df_unstacked = df_stacked.unstack()
df_unstacked
```

## ✅ Conclusión
Ahora puedes cambiar la forma de tus datos entre estructuras largas y anchas, y aplicar transformaciones más complejas con `stack()` y `unstack()` cuando trabajas con índices jerárquicos.
