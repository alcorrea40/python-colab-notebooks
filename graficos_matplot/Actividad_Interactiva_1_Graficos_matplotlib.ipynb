# ðŸ“Š Actividad Interactiva #1: GrÃ¡ficos con matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

## ðŸ”¢ Crear conjunto de datos

```python
x = np.arange(1, 6)
y = np.array([10, 15, 7, 12, 9])
```

## ðŸ“ˆ GrÃ¡fico de lÃ­neas

```python
plt.plot(x, y, marker='o')
plt.title('Ventas por dÃ­a')
plt.xlabel('DÃ­a')
plt.ylabel('Ventas')
plt.grid(True)
plt.show()
```

## ðŸ“Š GrÃ¡fico de barras

```python
plt.bar(x, y, color='skyblue')
plt.title('Ventas por dÃ­a')
plt.xlabel('DÃ­a')
plt.ylabel('Ventas')
plt.show()
```

## ðŸ”˜ GrÃ¡fico de dispersiÃ³n

```python
plt.scatter(x, y, color='red')
plt.title('RelaciÃ³n DÃ­a-Ventas')
plt.xlabel('DÃ­a')
plt.ylabel('Ventas')
plt.show()
```
