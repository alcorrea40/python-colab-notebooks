# 📊 Actividad Interactiva #1: Gráficos con matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

## 🔢 Crear conjunto de datos

```python
x = np.arange(1, 6)
y = np.array([10, 15, 7, 12, 9])
```

## 📈 Gráfico de líneas

```python
plt.plot(x, y, marker='o')
plt.title('Ventas por día')
plt.xlabel('Día')
plt.ylabel('Ventas')
plt.grid(True)
plt.show()
```

## 📊 Gráfico de barras

```python
plt.bar(x, y, color='skyblue')
plt.title('Ventas por día')
plt.xlabel('Día')
plt.ylabel('Ventas')
plt.show()
```

## 🔘 Gráfico de dispersión

```python
plt.scatter(x, y, color='red')
plt.title('Relación Día-Ventas')
plt.xlabel('Día')
plt.ylabel('Ventas')
plt.show()
```
