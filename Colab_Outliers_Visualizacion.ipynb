# Actividad Interactiva #1: Identificación de Outliers con Estadísticas y Visualización
En esta actividad aprenderás a detectar valores atípicos (outliers) en tus datos utilizando métodos como el rango intercuartílico (IQR), el z-score y representaciones gráficas como los boxplots.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

## 1. Crear un DataFrame de ejemplo

```python
np.random.seed(42)
valores = np.append(np.random.normal(loc=50, scale=10, size=100), [120, 130, 140])  # outliers
df = pd.DataFrame({'valor': valores})
df.head()
```

## 2. Visualizar datos con boxplot

```python
plt.figure(figsize=(8, 4))
sns.boxplot(x=df['valor'], color='skyblue')
plt.title('Boxplot de valores')
plt.grid(True)
plt.show()
```

## 3. Detectar outliers usando IQR

```python
Q1 = df['valor'].quantile(0.25)
Q3 = df['valor'].quantile(0.75)
IQR = Q3 - Q1
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR
outliers_iqr = df[(df['valor'] < limite_inferior) | (df['valor'] > limite_superior)]
outliers_iqr
```

## 4. Detectar outliers usando z-score

```python
df['zscore'] = stats.zscore(df['valor'])
outliers_z = df[np.abs(df['zscore']) > 3]
outliers_z
```

## ✅ Conclusión
