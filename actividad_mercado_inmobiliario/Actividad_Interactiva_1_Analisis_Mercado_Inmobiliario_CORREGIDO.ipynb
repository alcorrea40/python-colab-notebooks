## üìà An√°lisis univariado de variables clave

```python
sns.histplot(df['SalePrice'], kde=True)
plt.title('Distribuci√≥n del Precio de Venta')
plt.show()
```

```python
sns.histplot(df['GrLivArea'], kde=True)
plt.title('√Årea Habitable sobre el Nivel del Suelo')
plt.show()
```

## üîç An√°lisis bivariado

```python
sns.scatterplot(x='GrLivArea', y='SalePrice', data=df)
plt.title('Precio vs √Årea Habitable')
plt.show()
```

```python
sns.boxplot(x='Neighborhood', y='SalePrice', data=df)
plt.xticks(rotation=90)
plt.title('Precio por Vecindario')
plt.show()
```

## üîó Mapa de calor de correlaciones entre variables num√©ricas

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr[['SalePrice']].sort_values('SalePrice', ascending=False), annot=True, cmap='coolwarm')
plt.title('Correlaci√≥n con Precio de Venta')
plt.show()
```
