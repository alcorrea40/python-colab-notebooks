# Actividad Interactiva #2: Análisis temporal con resample() y rolling()
En esta actividad aprenderás a reagrupar datos de series temporales usando `resample()` y a suavizar tendencias mediante `rolling()` en pandas.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Crear una serie temporal de ejemplo

```python
np.random.seed(0)
fechas = pd.date_range(start='2022-01-01', periods=180, freq='D')
valores = np.random.normal(loc=50, scale=10, size=len(fechas))
df = pd.DataFrame({'fecha': fechas, 'valor': valores})
df.set_index('fecha', inplace=True)
df.head()
```

## 2. Visualizar los datos originales

```python
df.plot(figsize=(10,4), title='Datos originales')
plt.xlabel('Fecha')
plt.ylabel('Valor')
plt.grid(True)
plt.show()
```

## 3. Agrupar por mes con `resample('M')` y calcular media

```python
df_mensual = df.resample('M').mean()
df_mensual.plot(title='Media mensual')
plt.grid(True)
plt.show()
```

## 4. Aplicar promedio móvil con `rolling(window=7)`

```python
df['media_movil_7d'] = df['valor'].rolling(window=7).mean()
df[['valor', 'media_movil_7d']].plot(title='Promedio móvil 7 días')
plt.grid(True)
plt.show()
```

## ✅ Conclusión
Has aprendido a reagrupar datos por periodos de tiempo y a aplicar promedios móviles para suavizar la información y detectar patrones temporales más claros.
