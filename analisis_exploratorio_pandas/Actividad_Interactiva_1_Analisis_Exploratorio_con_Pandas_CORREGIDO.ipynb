# ðŸ§ª Actividad Interactiva #1: AnÃ¡lisis exploratorio completo con pandas

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## ðŸ“¥ Cargar dataset `titanic` o `diamonds`

```python
# Puedes cambiar 'titanic' por 'diamonds' si prefieres ese dataset
df = sns.load_dataset('titanic')
df.head()
```

## ðŸ§¹ Limpieza de datos

```python
df.info()
df.isnull().sum()
```

```python
# Eliminamos columnas con muchos nulos y reemplazamos valores faltantes
df = df.drop(columns=['deck'])
df['age'] = df['age'].fillna(df['age'].median())
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])
```

## ðŸ“Š AnÃ¡lisis univariado

```python
df.describe(include='all')
```

```python
sns.histplot(df['age'], kde=True)
plt.title('DistribuciÃ³n de Edad')
plt.show()
```

```python
sns.countplot(x='class', data=df)
plt.title('Frecuencia por Clase')
plt.show()
```

## ðŸ” AnÃ¡lisis bivariado

```python
sns.boxplot(x='class', y='age', data=df)
plt.title('Edad por Clase de Boleto')
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Mapa de Calor de CorrelaciÃ³n')
plt.show()
```
