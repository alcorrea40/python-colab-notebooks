```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

##  Cargar dataset de ejemplo (`titanic` de seaborn)

```python
df = sns.load_dataset('titanic')
```

## Ч Funci贸n de limpieza de datos

```python
def limpiar_datos(data):
    data = data.drop(columns=['deck'])
    data['age'] = data['age'].fillna(data['age'].median())
    data['embarked'] = data['embarked'].fillna(data['embarked'].mode()[0])
    return data
```

##  Funci贸n para mostrar estad铆sticas

```python
def mostrar_estadisticas(data):
    print("Resumen estad铆stico:")
    display(data.describe(include='all'))
    print("Valores nulos:")
    print(data.isnull().sum())
```

##  Funci贸n para generar visualizaciones

```python
def graficar_datos(data):
    sns.histplot(data['age'], kde=True)
    plt.title('Distribuci贸n de Edad')
    plt.show()

    sns.boxplot(x='class', y='age', data=data)
    plt.title('Edad por Clase de Boleto')
    plt.show()
```

##  Ejecutar flujo completo de an谩lisis

```python
df_limpio = limpiar_datos(df)
mostrar_estadisticas(df_limpio)
graficar_datos(df_limpio)
```
