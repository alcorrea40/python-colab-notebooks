# Actividad Interactiva #2: Manejo y transformación de variables categóricas
En este cuaderno aprenderás a trabajar con columnas categóricas usando pandas. Practicarás cómo convertir columnas a tipo `category`, modificar sus niveles y ordenarlas.

```python
import pandas as pd
import numpy as np
```

## 1. Crear un DataFrame con columnas categóricas

```python
df = pd.DataFrame({
    'nombre': ['Ana', 'Luis', 'Carlos', 'Ana', 'Luis'],
    'nivel': ['Básico', 'Avanzado', 'Intermedio', 'Avanzado', 'Básico']
})
df
```

## 2. Convertir la columna 'nivel' a tipo categórico

```python
df['nivel'] = df['nivel'].astype('category')
df.dtypes
```

## 3. Redefinir y ordenar categorías

```python
df['nivel'] = df['nivel'].cat.set_categories(['Básico', 'Intermedio', 'Avanzado'], ordered=True)
df['nivel']
```

## 4. Codificar las categorías con `cat.codes`

```python
df['nivel_codigo'] = df['nivel'].cat.codes
df
```

## 5. Visualizar y comparar la codificación

```python
print("Valores únicos en 'nivel':", df['nivel'].unique())
print("Códigos asignados:", df['nivel_codigo'].unique())
```
