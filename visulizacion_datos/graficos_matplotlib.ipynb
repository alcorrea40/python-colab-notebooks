# 游늵 Visualizaci칩n de datos con matplotlib
Este notebook te ense침a a crear gr치ficos de l칤neas, barras y dispersi칩n usando `matplotlib.pyplot`, personalizando t칤tulos, etiquetas y estilos.

```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

# Subir archivo CSV
uploaded = files.upload()

# Leer archivo
df = pd.read_csv("datos_graficos.csv")
df
```

## 游늳 Gr치fico de l칤neas

```python
plt.plot(df['Mes'], df['Ventas'], marker='o', label='Ventas')
plt.plot(df['Mes'], df['Gastos'], marker='s', label='Gastos')
plt.title('Ventas y Gastos Mensuales')
plt.xlabel('Mes')
plt.ylabel('Valor en USD')
plt.legend()
plt.grid(True)
plt.show()
```

## 游늵 Gr치fico de barras

```python
plt.bar(df['Mes'], df['Ventas'], color='skyblue')
plt.title('Ventas por Mes')
plt.xlabel('Mes')
plt.ylabel('Ventas (USD)')
plt.xticks(rotation=45)
plt.show()
```

## 游댯 Gr치fico de dispersi칩n

```python
plt.scatter(df['Ventas'], df['Gastos'], color='red')
plt.title('Relaci칩n entre Ventas y Gastos')
plt.xlabel('Ventas')
plt.ylabel('Gastos')
plt.grid(True)
plt.show()
```

## 游빍 Ejercicio pr치ctico

```python
# Crea un gr치fico combinado de barras para ventas y gastos
import numpy as np
ancho = 0.35
x = np.arange(len(df['Mes']))

plt.bar(x - ancho/2, df['Ventas'], width=ancho, label='Ventas')
plt.bar(x + ancho/2, df['Gastos'], width=ancho, label='Gastos')

plt.xlabel('Mes')
plt.ylabel('USD')
plt.title('Comparaci칩n de Ventas y Gastos')
plt.xticks(x, df['Mes'], rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
