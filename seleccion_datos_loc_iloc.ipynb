# üìì Selecci√≥n de datos con `loc[]` e `iloc[]` en pandas
Este notebook te ense√±a a:
- Seleccionar columnas y filas espec√≠ficas
- Filtrar por condiciones
- Usar `loc[]` para etiquetas y `iloc[]` para posiciones

```python
import pandas as pd
from google.colab import files

# Subir el archivo empleados.csv
uploaded = files.upload()
df = pd.read_csv("empleados.csv")

print("Archivo cargado correctamente.")
df.head()
```

## Selecci√≥n de columnas espec√≠ficas

```python
# Una sola columna
print(df["Nombre"])

# Varias columnas
print(df[["Nombre", "Ciudad"]])
```

## Uso de `loc[]` ‚Äì Selecci√≥n por etiquetas

```python
# Filtrar por condici√≥n y seleccionar columnas
bogota = df.loc[df["Ciudad"] == "Bogot√°", ["Nombre", "Salario"]]
print(bogota)

# Filas por rango de √≠ndices
print(df.loc[1:3])
```

## Uso de `iloc[]` ‚Äì Selecci√≥n por posici√≥n

```python
# Fila individual
print(df.iloc[2])

# Rango de filas y columnas
print(df.iloc[0:3, 0:2])
```

## üß™ Ejercicios pr√°cticos

```python
# 1. Nombres y edades con salario > 2500
print(df.loc[df["Salario"] > 2500, ["Nombre", "Edad"]])

# 2. √öltimas 2 filas con iloc
print(df.iloc[-2:])

# 3. Filas 0, 2 y 4 y columnas 'Nombre' y 'Ciudad'
print(df.loc[[0, 2, 4], ["Nombre", "Ciudad"]])
```
