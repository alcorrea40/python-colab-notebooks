# 游늽 Funciones estad칤sticas b치sicas en pandas
Este notebook te ense침a a aplicar funciones estad칤sticas b치sicas como `mean()`, `median()`, `std()`, `describe()`, etc., sobre columnas num칠ricas de un DataFrame.

```python
import pandas as pd
from google.colab import files

# Subir archivo CSV
uploaded = files.upload()

# Leer archivo
df = pd.read_csv("estadisticas_dataframe.csv")
df
```

## 游늵 Descripci칩n general de los datos

```python
# Resumen estad칤stico general
df.describe()
```

## 游댝 Medidas de tendencia central

```python
# Media y mediana
print("Media de Edad:", df["Edad"].mean())
print("Mediana de Salario:", df["Salario"].median())
```

## 游늳 Medidas de dispersi칩n

```python
# Desviaci칩n est치ndar, varianza y rango
print("Desviaci칩n est치ndar de Edad:", df["Edad"].std())
print("Varianza de Salario:", df["Salario"].var())
print("Rango de HorasTrabajoSemanal:", df["HorasTrabajoSemanal"].max() - df["HorasTrabajoSemanal"].min())
```

## 游빍 Ejercicios pr치cticos

```python
# 1. Valor m치ximo y m칤nimo del salario
print("Salario m치ximo:", df["Salario"].max())
print("Salario m칤nimo:", df["Salario"].min())

# 2. Media y desviaci칩n est치ndar de horas trabajadas
print("Media de horas trabajadas:", df["HorasTrabajoSemanal"].mean())
print("Desviaci칩n est치ndar de horas trabajadas:", df["HorasTrabajoSemanal"].std())
```
