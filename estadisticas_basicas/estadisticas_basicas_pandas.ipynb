# üìì Funciones estad√≠sticas b√°sicas en pandas
Este notebook te ense√±a a aplicar funciones estad√≠sticas b√°sicas como `mean()`, `median()`, `std()`, `describe()`, etc., sobre columnas num√©ricas de un DataFrame.

```python
import pandas as pd
from google.colab import files

# Subir archivo CSV
uploaded = files.upload()

# Leer archivo
df = pd.read_csv("estadisticas_dataframe.csv")
df
```

## üìä Descripci√≥n general de los datos

```python
# Resumen estad√≠stico general
df.describe()
```

## üî¢ Medidas de tendencia central

```python
# Media y mediana
print("Media de Edad:", df["Edad"].mean())
print("Mediana de Salario:", df["Salario"].median())
```

## üìà Medidas de dispersi√≥n

```python
# Desviaci√≥n est√°ndar, varianza y rango
print("Desviaci√≥n est√°ndar de Edad:", df["Edad"].std())
print("Varianza de Salario:", df["Salario"].var())
print("Rango de HorasTrabajoSemanal:", df["HorasTrabajoSemanal"].max() - df["HorasTrabajoSemanal"].min())
```

## üß™ Ejercicios pr√°cticos

```python
# 1. Valor m√°ximo y m√≠nimo del salario
print("Salario m√°ximo:", df["Salario"].max())
print("Salario m√≠nimo:", df["Salario"].min())

# 2. Media y desviaci√≥n est√°ndar de horas trabajadas
print("Media de horas trabajadas:", df["HorasTrabajoSemanal"].mean())
print("Desviaci√≥n est√°ndar de horas trabajadas:", df["HorasTrabajoSemanal"].std())
```
